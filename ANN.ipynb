```python
import pandas as pd
import numpy as np 
import seaborn as sns
import warnings 
warnings.warn('ignore')
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('breast_cancer11.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['diagnosis'] = data['diagnosis'].replace({'M':1, 'B':0})
```

```python
data.head()
```

```python
data.drop(columns=['id'], inplace=True)
```

```python
data.head()
```

```python
data = data.drop(columns=['Unnamed: 32'])
data.head()
```

```python
corr_matrix = data.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.show()
```

```python
X = data.drop(columns=['diagnosis'])
y = data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train.head()
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
main_features = ["radius_mean", "texture_mean", "smoothness_mean", "concavity_mean", "symmetry_mean"]
data_features = data[main_features + ["diagnosis"]]
```

```python
data_features.head()
```

```python
corr_matrix = data_features.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm")
plt.show()
```

```python
figures={}
fig, ax=plt.subplots(figsize=(15, 12))
data_features.hist(ax=ax, bins=20, edgecolor="black")
figures["histograms"] = fig
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled_x=sc.fit_transform(X)
X = pd.DataFrame(scaled_x, columns=X.columns)
X.head()
```

```python
X
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier( hidden_layer_sizes=(50,3),
                       learning_rate_init=0.1,
                       max_iter=100,
                       random_state=2) 
model.fit(X_train,y_train) 
y_predict_proba = model.predict_proba(X_test) 
y_predict = model.predict(X_test)
y_train_predict = model.predict(X_train)
```

```python
y_predict_proba
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print("Train accuracy :",accuracy_score(y_train,y_train_predict))
print("Test accuracy :",accuracy_score(y_test,y_predict))
```
